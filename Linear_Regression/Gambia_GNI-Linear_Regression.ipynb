### We are going to work with a Linear Regression Model to learn the relationship between the columns in our data and then Predict the Gross National Income of the Gambia in the coming Years.

```python
# Importing the neccesary libraries that we will be working with
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
from sklearn.linear_model import LinearRegression
```

```python
# Using pandas to read in out dataset and only use our specified colums
data = pd.read_csv('Gambia_Gross-National-Income.csv',usecols=('Year','GNI'))
```

```python
# Taking a look at the top hows in our dataset
data.head()
```

```python
'''Getting summary of the data, including the number of rows and columns, the 
data types of each column, and the number of non-null values in each column'''
data.info()
```

```python
# Using the searborn regplot method to visualize the relations between our data
plt.style.use('bmh')
plt.figure(figsize=(10,8))
plt.xticks(rotation=45)
sns.regplot(x='Year' ,y = 'GNI', data=data,color='green')
```

```python
'''Creating  a linear regression model using the LinearRegression class from scikit-learn to train our model. 
It fits the model to the data, using the Year column as the 
independent variable and the GNI column as the dependent variable.'''
model = LinearRegression()
model.fit(data[['Year']].values, data.GNI)
```

```python
# Predict the GNI for the year 2019 using the trained linear regression model
prediction = model.predict([[2019]])
prediction
```

```python
# Predict the GNI for the year 2020 using the trained linear regression model
prediction = model.predict([[2020]])
prediction
```
